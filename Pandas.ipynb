```python
##handling NaN values 
#1>we directly remove them
#2>or we replace with some values
```

```python
import numpy as np
df=iris.copy()
df.columns = ['sl', 'sw', 'pl', 'pw', 'flower_type']
df
```

```python
##adding some nan vaues with the help of numpy
df.iloc[2:4, 1:3] = np.nan
df.head()
```

```python
## dropna()-this will remove row or column entries with nan values
df.dropna(inplace = True)  ## Remove NaN inside df only
df.head()
```

```python
df.reset_index(drop = True, inplace = True)   ## Reset the indices
```

```python
df.head()
```

```python
##filling nan value(fillna()-used to fill nan values)
df.iloc[2:4, 1:3] = np.nan
df.head()
```

```python
df.sw.fillna(df.sw.mean(), inplace = True)
df.pl.fillna(df.pl.mean(), inplace = True)
df.head()
```

```python
## we can see the nan values generally belong to iris-setosa 
df_setosa = df[df.flower_type == 'Iris-setosa']
df.sw.fillna(df_setosa.sw.mean(), inplace = True)
df.pl.fillna(df_setosa.pl.mean(), inplace = True)
df.head()
```
