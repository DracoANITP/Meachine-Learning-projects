# plotting graphs
### -scatter graph, line graph , adding styles, different types of plot

```python
## importing the graph library
import matplotlib.pyplot as plt
import numpy as np
```

```python
##plotting the line graph
x=np.array([1,6,3,8,9,11])
plt.plot(x)
plt.show()
```

```python
## now we use scatter
x=np.array([1,6,3,8,9,11])
y=np.array([3,9,4,1,10,2])
plt.scatter(x,y)
plt.show()
```

```python
# Plotting a function with respect to given points. The 'ro' 
# parameter specifies two things: the 'r' specifies the color of 
# the points which is red, and the 'o' specifies the shape.
x = np.array([1, 2, 3, 4, 5])
y = x**4
plt.plot(x, y, 'ro')
plt.show()
```

```python
##plotting line graph y=mx+b
```

```python
m = 10
b = 5
X = np.array([1,8,20,45])
Y = m*X + b
plt.plot(X,Y)
plt.show()
```

```python
X = np.array([1,2,4,5])
Y = np.array([10,35,78,10])

# Color property can be used to change colours

plt.plot(X,Y,color = 'orange')
# We can use any color here in values

plt.show()
```

```python
X = np.array([1,2,4,5])
Y = np.array([10,35,78,10])

# Marker property can be used to change markers to any valid marker style like circle,diamond,triangle,etc.

plt.plot(X,Y,marker = 'o')
# We can use any color here in values

plt.show()
```

```python
X = np.array([1,2,3,4])
Y = np.array([13,9,78,45])
plt.plot(X,Y,linewidth = 4.5)
# Linewidth can take any float values
plt.show()
```

```python
##plotting multi values
t = np.arange(0, 5, 0.2)
# t has evenly sampled values from 0 to 5 with 0.2 step value
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
# red dashes, blue squares and green triangles
# we can plot multiple equations on the same figure
plt.show()
```

# adding labels and title to the graph

```python
X = np.array([1,2,3,4])
Y = np.array([13,9,78,45])
plt.plot(X,Y)
plt.ylabel("Some random values")
plt.xlabel("Points")
# Adding x and y axis labels
plt.title("Plotting random values v/s data points")
# Adding title on the graphs
plt.show()
```

```python
plt.plot([0.1, 0.2, 0.3, 0.4], [1, 2, 3, 4], label='first plot')
plt.plot([0.1, 0.2, 0.3, 0.4], [1, 4, 9, 16], label='second plot')
plt.legend()
plt.show()
```

# plotting pie graph

```python
pieLabels = ['Asia', 'Africa', 'Europe', 'North America', 'South America']
populationShare = [59.69, 16, 9.94, 7.79, 5.68]

figureObject, axesObject = plt.subplots()

axesObject.pie(populationShare,labels=pieLabels,autopct='%.1f%%',startangle=90, wedgeprops = {'linewidth': 5})
axesObject.axis('equal')
plt.show()
```

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  
# only "explode" the 2nd slice (i.e. 'Hogs')

plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

# plotting bubble graph


```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (15 * np.random.rand(N))**2

# s - size of points, can be a single number or an array
# c - colour of the points
# alpha - transparency value of colours
plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
```
